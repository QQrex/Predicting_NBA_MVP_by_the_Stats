```python
# call teams to extract team id
from nba_api.stats.static import teams
# get_teams returns a list of 30 dictionaries, each an NBA team.
nba_teams = teams.get_teams()
print('Number of teams fetched: {}'.format(len(nba_teams)))
nba_teams
```

```python
# call players
from nba_api.stats.static import players
# get_players returns a list of dictionaries, each representing a player.
nba_players = players.get_players()
print('Number of players fetched: {}'.format(len(nba_players)))
nba_players
```

```python
# json_normalize teams to df
team_df = pd.json_normalize(nba_teams)

team_df
```

```python
# json_normalize palyers to df
nba_players_df = pd.json_normalize(nba_players)

nba_players_df
```

```python
# Create test list for call to api
game_list_test = game_id_list[0:20]

game_list_test
```

```python
# Create empty DataFrame

all_games_df_2122 = pd.DataFrame(columns = ['GAME_ID', 'TEAM_ID', 'TEAM_ABBREVIATION', 'TEAM_CITY', 
                                            'PLAYER_ID', 'PLAYER_NAME','MIN', 'MIN', 'FGM','FGA', 
                                            'FG_PCT', 'FG3M', 'FG3A', 'FG3_PCT', 'FTM', 'FTA', 
                                            'FT_PCT', 'OREB', 'DREB', 'REB', 'AST', 'STL', 'BLK', 
                                            'TO', 'PF', 'PTS', 'PLUS_MINUS'])

# create counter
record_counter = 1
set_counter = 1
```

```python
# test for loop to call scoreboard data with list of game_id
for i, gameid in enumerate(game_list_test):
    
    if (i % 5 == 0 and i >= 5):
        set_counter += 1
        record_count = 1
        time.sleep(10)
    
    # print status
    print(f'Processing record {record_counter} of Set {set_counter}')
    
    record_counter += 1
    
    game_test = boxscoretraditionalv2.BoxScoreTraditionalV2(game_id= gameid)
    game_test_df = game_test.get_data_frames()[0][['GAME_ID', 'TEAM_ID', 'TEAM_ABBREVIATION', 'TEAM_CITY', 
                                                    'PLAYER_ID', 'PLAYER_NAME','MIN', 'MIN', 'FGM','FGA', 
                                                    'FG_PCT', 'FG3M', 'FG3A', 'FG3_PCT', 'FTM', 'FTA', 
                                                    'FT_PCT', 'OREB', 'DREB', 'REB', 'AST', 'STL', 'BLK', 
                                                    'TO', 'PF', 'PTS', 'PLUS_MINUS']]
    all_games_df_2122 = all_games_df_2122.append(game_test_df)
    
print(f'complete')
```
